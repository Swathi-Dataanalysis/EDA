Statistical Tests for Categorical Variables:

For relationships involving categorical variables (like gender, item purchased, subscription status), we are using chi-square tests for independence or t-tests/ANOVA for comparing means across groups.

Chi-Square Test: To test if there's a significant association between two categorical variables (e.g., gender and product preference).
T-Test/ANOVA: To compare the means of a numerical variable across different categories (e.g., average purchase amount across different genders).

```python
from scipy import stats

# Example: Comparing average purchase amount between two genders
group1 = df[df['Gender'] == 'Male']['Purchase Amount (USD)']
group2 = df[df['Gender'] == 'Female']['Purchase Amount (USD)']

# Independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05

if p < alpha:
    print("We reject the null hypothesis. There is a significant difference between average purchase amount between two genders")
else:
    print("We do not reject the null hypothesis. There is no significant difference between average purchase amount between two genders")
```

2. Gender and Product Preference
Hypothesis: Male and female customers show distinct preferences for certain categories of products.

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Creating a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Item Purchased'])

# Performing the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-value:", p)

# Interpretation
alpha = 0.05

if p < alpha:
    print("We reject the null hypothesis. There is a significant association between gender and item purchased.")
else:
    print("We do not reject the null hypothesis. There is no significant association between gender and item purchased.")
```

3. Review Ratings and Repeat Purchases
Hypothesis: Higher review ratings are associated with increased frequency of purchases.
Review Rating" across different categories of "Frequency of Purchases" using ANOVA 

```python
from scipy.stats import f_oneway

# We have categories like 'Fortnightly', 'Bi-Weekly', 'Weekly', 'Monthly', 'Quarterly', 'Annually' in 'Frequency of Purchases'
groups = df['Frequency of Purchases'].unique()
group_data = [df[df['Frequency of Purchases'] == group]['Review Rating'] for group in groups]

f_stat, p_value = f_oneway(*group_data)
print("F-Statistic:", f_stat)
print("P-Value:", p_value)
#Interpretation
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference in review ratings across different frequencies of purchases.")
else:
    print("We do not reject the null hypothesis. There is no significant difference in review ratings across different frequencies of purchases.")
```

4. Impact of Discounts and Promotions
Hypothesis: Customers who use promo codes or receive discounts tend to make larger purchases.

```python
from scipy.stats import ttest_ind

# Assuming 'Purchase Amount (USD)' is your numerical variable
group1 = df[df['Discount Applied'] == 'Yes']['Purchase Amount (USD)']
group2 = df[df['Discount Applied'] == 'No']['Purchase Amount (USD)']

# Independent t-test
t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')  # 'nan_policy' is set to 'omit' to handle any NaNs

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference in purchase amounts between customers who received discounts and those who did not.")
else:
    print("We do not reject the null hypothesis. There is no significant difference in purchase amounts between customers who received discounts and those who did not.")
```

```python
from scipy.stats import ttest_ind

# Assuming 'Purchase Amount (USD)' is your numerical variable
group1 = df[df['Promo Code Used'] == 'Yes']['Purchase Amount (USD)']
group2 = df[df['Promo Code Used'] == 'No']['Purchase Amount (USD)']

# Independent t-test
t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')  # 'nan_policy' is set to 'omit' to handle any NaNs

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference in purchase amounts between customers who received discounts and those who did not.")
else:
    print("We do not reject the null hypothesis. There is no significant difference in purchase amounts between customers who received discounts and those who did not.")
```

5. Subscription Status and Customer Loyalty
Hypothesis: Subscription status correlates with higher spending or more frequent purchases.

```python
#Using t-test
group1 = df[df['Subscription Status'] == 'Yes']['Purchase Amount (USD)']
group2 = df[df['Subscription Status'] == 'No']['Purchase Amount (USD)']

t_stat, p_value = ttest_ind(group1, group2)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference in purchase amounts between customers with and without a subscription.")
else:
    print("We do not reject the null hypothesis. There is no significant difference in purchase amounts between customers with and without a subscription.")
```

6. Seasonal Variations in Shopping Trends
Hypothesis: Certain seasons are associated with higher spending or specific types of purchases.

```python
# One-way ANOVA
f_stat, p_value = f_oneway(df[df['Season'] == 'Spring']['Purchase Amount (USD)'],
                           df[df['Season'] == 'Summer']['Purchase Amount (USD)'],
                           df[df['Season'] == 'Fall']['Purchase Amount (USD)'],
                           df[df['Season'] == 'Winter']['Purchase Amount (USD)'])
print("F-Statistic:", f_stat)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05

if p_value < alpha:
    print("We reject the null hypothesis. There are significant differences in purchase amounts across different seasons.")
else:
    print("We do not reject the null hypothesis. There are no significant differences in purchase amounts across different seasons.")
```

Tukey's Honestly Significant Difference (HSD) test. This test compares all possible pairs of means to determine which seasons are significantly different from each other.

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.api as sm

# Assuming df is your DataFrame and you've already performed ANOVA

# Prepare the data for Tukey's test
tukey_data = df[['Season', 'Purchase Amount (USD)']].dropna()

# Tukey's HSD test
tukey_results = pairwise_tukeyhsd(endog=tukey_data['Purchase Amount (USD)'], groups=tukey_data['Season'], alpha=0.05)

# Print the results
print(tukey_results)

# Plot the results
tukey_results.plot_simultaneous(xlabel='Mean Purchase Amount (USD)', ylabel='Season')
plt.show()
```

There is a sgnificant differences in purchase amounts across the seasonsFall and season spring .
